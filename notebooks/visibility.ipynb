### StormSight Data Exploration
###### **Analyzing rainy text image datasets**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

### Configuration

```python
DATA_DIR = Path("../data/")
RAINY_IMAGES = list(DATA_DIR.glob("inputs/*.jpg"))
```

```python
def plot_comparison(rainy_path):
    fig, ax1 = plt.subplots(1, 1, figsize=(12, 6))
    
    rainy_img = cv2.cvtColor(cv2.imread(str(rainy_path)), cv2.COLOR_BGR2RGB)
    
    ax1.imshow(rainy_img)
    ax1.set_title(f"Rainy Image\n{rainy_path.name}")
    plt.show()
```

### Visualize sample images

```python
for rainy in RAINY_IMAGES[:3]:
    plot_comparison(rainy)
```

### Text Visibility Metrics

```python
def calculate_visibility(img_path):
    img = cv2.imread(str(img_path))
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Edge-based visibility score
    edges = cv2.Canny(gray, 100, 200)
    visibility = np.sum(edges) / (img.shape[0] * img.shape[1])
    
    return visibility
```

### Compare visibility scores

```python
for img in RAINY_IMAGES[:5]:
    print(f"{img.name}: {calculate_visibility(img):.2f} visibility score")
```
